```python
import pysam
import pandas as pd
import numpy as np 
import os 
import sys 
```

```python

input_anno = pysam.TabixFile("../../annotations/gerp/gerp_score2_hg38_MAM_90q.bed.gz")
input_bed = pd.read_table("../../beds/nstd175/nstd175vcf_nochr.bed", header=None) 
```

```python
for i, row in input_bed.iterrows():
    window = []
    for gc in input_anno.fetch(row[0], row[1], row[2], parser=pysam.asBed()):
        window.append(float(gc[3]))
    if window == []:
      print(row[0], row[1], row[2], ".", sep="\t")
    else:
      print(row[0], row[1], row[2], max(window), sep="\t")
```
